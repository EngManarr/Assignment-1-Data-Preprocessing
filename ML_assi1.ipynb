```python
# Memory Management
def optimize_memory(df):
    for col in df.columns:
        if df[col].dtype == 'float64':
            df[col] = pd.to_numeric(df[col], downcast='float')
        elif df[col].dtype == 'int64':
            df[col] = pd.to_numeric(df[col], downcast='integer')
        elif df[col].dtype == 'object':
            if df[col].nunique() < 50:
                df[col] = df[col].astype('category')
    return df

train_df = optimize_memory(df)
test_df = optimize_memory(df2)
print("\nMemory optimized successfully!")
```

```python
import pandas as pd
import numpy as np

# أخذ نسخة جديدة
train_df_clean = df.copy()
test_df_clean = df2.copy()

print(" تنظيف البيانات")

# حذف الأعمد الأكثر فقداناً
columns_to_drop = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu']
train_df_clean = train_df_clean.drop(columns=columns_to_drop, errors='ignore')
test_df_clean = test_df_clean.drop(columns=columns_to_drop, errors='ignore')

print(" تم حذف الأعمدة بنجاح")



#  معالجة الأعمدة الفئوية
# الأعمدة النصية التي تعني فيها NaNلا يوجد ميزة في المنزل نضع None
cat_structural_cols = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
                        'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType']

for col in cat_structural_cols:
    for df in [train_df_clean, test_df_clean]:
        if col in df.columns:
            if df[col].dtype.name == 'category':
                if 'None' not in df[col].cat.categories:
                    df[col] = df[col].cat.add_categories(['None'])
                df[col] = df[col].fillna('None')
            else:
                df[col] = df[col].fillna('None')


#  معالجة الميزات الرقمية
# إذا كان لا يوجد كراج أو قبو فإن الأعمدة الرقمية الخاصة بهما تكون 0
zero_numerical_cols = ['GarageYrBlt', 'GarageCars', 'GarageArea', 'MasVnrArea',
                       'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF',
                       'BsmtFullBath', 'BsmtHalfBath']

for col in zero_numerical_cols:
    if col in train_df_clean.columns:
        train_df_clean[col] = train_df_clean[col].fillna(0)
    if col in test_df_clean.columns:
        test_df_clean[col] = test_df_clean[col].fillna(0)


# معالجة عمود واجهة الأرض الرقمي (LotFrontage)
# نملأ واجهة الأرض بالوسيط لكل حي سكني على حدة من ملف التدريب لضمان الدقة وتجنب تسريب البيانات
train_df_clean['LotFrontage'] = train_df_clean.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))
test_df_clean['LotFrontage'] = test_df_clean.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(train_df_clean['LotFrontage'].median()))


# لتنظيف الميزات النادرة في الطرفين
for col in train_df_clean.columns:

    #  تعبئة أي قيمة مفقودة متبقية في ملف التدريب
    if train_df_clean[col].isnull().sum() > 0:
        if train_df_clean[col].dtype.name in ['category', 'object']:
            train_df_clean[col] = train_df_clean[col].fillna(train_df_clean[col].mode()[0])
        else:
            train_df_clean[col] = train_df_clean[col].fillna(train_df_clean[col].median())

    # تعبئة ملف الاختبار بناءً على قيم ملف التدريب
    if col in test_df_clean.columns and test_df_clean[col].isnull().sum() > 0:
        if train_df_clean[col].dtype.name in ['category', 'object']:
            test_df_clean[col] = test_df_clean[col].fillna(train_df_clean[col].mode()[0])
        else:
            test_df_clean[col] = test_df_clean[col].fillna(train_df_clean[col].median())
```

```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

# تحديد مجموعات الأعمدة بناءً على نوع المعالجة
num_cols = ['GrLivArea', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'GarageCars', 'GarageArea'] # أضيفي باقي الأعمدة الرقمية هنا
ordinal_cols = ['ExterQual', 'KitchenQual']
nominal_cols = ['Neighborhood', 'BldgType']

# تحديد الترتيب المنطقي الدقيق للأعمدة الترتيبية (من السيئ إلى الممتاز)
# الترتيب يمرر كقائمة تحتوي على قائمة مستويات كل عمود بالترتيب التصاعدي
qual_order = ['Po', 'Fa', 'TA', 'Gd', 'Ex']
ordinal_categories = [qual_order, qual_order] # ترتيب متطابق للعمودين

#بناء خطوط المعالجة الفرعية لكل نوع من البيانات (Sub Pipelines)

#Pipeline البيانات الرقمية
num_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')) # تنظيف وتعبئة بالوسيط
])

#Pipeline البيانات الترتيبية
ordinal_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='TA')), # تعبئة الفراغات بالقيمة المتوسطة
])

#Pipeline البيانات الاسمية
nominal_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')), # تعبئة بالقيمة الأكثر تكراراً (Mode)
])

#دمج جميع العمليات داخل كائن ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_cols),
        ('ord', ordinal_pipeline, ordinal_cols),
        ('nom', nominal_pipeline, nominal_cols)
    ],
    remainder='drop' # استبعاد أي أعمدة أخرى لم نذكرها زي Id
)

# تطبيق البب لاين
full_pipeline = Pipeline(steps=[('preprocessor', preprocessor)])


# أولاً: نقوم بتجهيز المتغير المستهدف وتطبيق التحويل اللوغاريتمي عليه يدويًا
if 'SalePrice' in train_df.columns:
    y_train = np.log1p(train_df['SalePrice'])
    X_train_raw = train_df.drop(columns=['SalePrice'])
else:
    X_train_raw = train_df.copy()

#Pipeline تحويل بيانات التدريب
X_train_processed = full_pipeline.fit_transform(X_train_raw)

#تحويل بيانات الاختبار
X_test_processed = full_pipeline.transform(test_df)

print("تم بناء وتطبيق  Pipeline")
print(f"  التدريب : {X_train_processed.shape}")
print(f"  الاختبار : {X_test_processed.shape}")
```
